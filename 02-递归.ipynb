```python
# 递归实现连加,类似于栈 
def digui(num):
    print(num)
    if num > 0:
        digui(num-1)
    else:
        print("")
    print(num)

digui(3)
```

```python
# 汉诺塔递归问题
i = 0
def move(n, a, b, c):
    global i
    if n == 1:
        i += 1
        print("移动第",i,"次",a,'-->',c)
    else:
        move(n-1,a,c,b)  #　借助ｃ将上面n-1个移到ｂ上
        move(1,a,b,c)    # 将ａ上最下面一个移到ｃ上
        move(n-1,b,a,c)  # 将ｂ上的ｎ-1个借助a移到ｃ
        
move(5,'A','B','C')
```

```python
# 利用递归获取文件夹下所有文件大小
import os
def getdirpath(dirpath):
    total = 0
    allname = os.listdir(dirpath)
    for name in allname:
        fullpath = os.path.join(dirpath, name) #　拼接成完整路径
        # print(fullpath)
        if os.path.isfile(fullpath):
            total += os.path.getsize(fullpath)
        elif os.path.isdir(fullpath):
            total += getdirpath(fullpath)
    return total

size = getdirpath('//home//tlxy//pythonstudy')
print(size)
```
