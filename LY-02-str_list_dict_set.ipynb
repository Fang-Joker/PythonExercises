```python
# 常用的操作
# append():追加,不生成新对象   
# 这些地方要熟悉清楚操作后是生成新对象，还是在原对象修改
# 可以通过id()判断
# copy(): 浅拷贝，生成一个新对象
# account():统计某元素的数量
# extend():追加一个新列表，直接在操作list上追加，函数本身返回为None，不生成新对象
list1 = [1,2,3]
print(id(list1))
list2 = [4,5,6]
list3 = list1.extend(list2)
# 在list1上直接加上了
print(list1)
print(list2)
print(list3)
# 看看是否生成新对象，实际是没有
print(id(list1))
```

```python
# insert():指定位置插入元素
l1 = [1,2,3]
l1.insert(1,10)
print(l1)
```

# Dict
- clear():清楚整个字典，返回None

```python
# fromkeys():按照指定序列为键创建字典，值都一样，可给出，不给则为None  返回dict
l1 = ["a","b"]
d1 = {}.fromkeys(l1)
d2 = {}.fromkeys(l1,3)
print(d1)
print(d2)
```

```python
# get():用键获取值，返回值；若不存在返回None,可以额外设置
# 比直接使用索引好，使用索引如果没有这个键会报错
# setdefault():搜索某键值对存在则返回值，没有时直接加入进去
dict1 = {"as":1}
print(dict1.get("as"))
print(dict1.get("asd"))
print(dict1.get("asd",-1))
```

```python
# items():将字典转换为元组,单个元素为键和值，就可以同个两个变量访问
a = {"a":1,"b":2}
print(a.items())
for i in a.items():
    print(i)
print("-"*20)
for k,v in a.items():
    print(k,v)
```

```python
# pop():移除指定键的键值对，如果不存在，则返回指定值；未设置指定值则报错
# popitem():直接移除一个键值对，无需指定键
a = {"a":1,"b":2,"c":3}
b = a.pop("a")
print(b)
print(a.pop("b"))
print(a.pop("f",-1))
print(a.pop("k"))
```
